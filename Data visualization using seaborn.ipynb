# Heatmap

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
x
```

```python
sns.heatmap(x, cbar=False,linewidths=0.5, annot=True,fmt='d',cmap='viridis')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='viridis')
```

# Clustermap


```python
sns.clustermap(x, z_score=0,row_cluster=True, metric='correlation')
```

```python
sns.clustermap(x, z_score=0,row_cluster=False,annot=True,metric='euclidean')
```

```python
sns.clustermap(x,z_score=0, row_cluster=False, annot=True, metric='correlation')
```
